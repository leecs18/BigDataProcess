```python
#9*9 multiple table
#version 1
for x in range(1,10):
  for y in range(1,10):
    print('%d*%d=%2d'%(x,y,x*y))
  print()
```

```python
print('=====')
for x in range(1,6):
  print('x = %d'%(x))
  for y in range(1,6):
    print(' y = %d'%(y))
  print('=====')
```

```python
#version 2
for x in range(1,10):
  for y in range(1,10):
    print('%d*%d=%2d'%(x,y,x*y),end = ' ')
  print()
```

```python
#version 3
for x in range(1,10):
  for y in range(1,10):
    print('%d*%d=%2d'%(y,x,x*y),end = ' ')
  print()
```

```python
#version 3
for x in range(1,10):
  for y in range(1,10):
    print(f'{x}*{y}={x*y:2d}',end=' ')
  print()
```
